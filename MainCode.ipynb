```python
def CalculateFinalNotes(barNotes, doneBars, codes, finalNotes):

  #Entering all the notes to make the code
  for i, focusNote in enumerate(barNotes[doneBars]):
    #print(focusNote)
    #Message('note_on', channel=1, note=67, velocity=100, time=960),
    #Make sure there is at least one musical note
    if focusNote.velocity != 0 or focusNote.type != 'note_on' or focusNote.time != 0:
      focusTime = 0
      #Enters if the note is on, so the midi starts giving signal
      if focusNote.velocity != 0 and focusNote.type != 'note_off':
        #As the first part of the code is the first note of the bar, if the code is empty, in the first note_on it is added to the code
        if codes[doneBars] == []:
          codes[doneBars].append(focusNote.note)
        #Entering a loop to get how much time does each note last
        for j, noteSearched in enumerate(barNotes[doneBars]):
          if j>i:
            #Adding time for each event that is not the note_off or the note_velocity == 0
            if focusNote.note != noteSearched.note:
              focusTime = focusTime + noteSearched.time
            #Once the note is over, it is added to "finalNotes" with the amount of time this note code has been played in the time space
            else:
              focusTime = focusTime + noteSearched.time
              if focusNote.note not in finalNotes:
                finalNotes[focusNote.note] = 0
              finalNotes[focusNote.note] = finalNotes[focusNote.note] + focusTime
              break
  return finalNotes, codes
```

```python
def DetectCode(midi, track=0, bars = 1, timeSignature = 4):
  beat = midi.ticks_per_beat
  #print(beat)
  bar = beat * timeSignature
  notes = list(midi.tracks[track])
  time = 0
  barNotes = [[]]
  doneBars = 0
  finalNotes = {}
  codedSong = []
  codes = [[]]
  ending = False
  #Entering in all the notes of the midi file
  for note in notes:
    unfinished = -1
    if ending and (note.type == 'note_on' or note.type == 'note_off'):
      ending = False
    #Adding a new set of notes for each extra bar
    if len(barNotes) <= doneBars + 1:
      barNotes.append([])
      codes.append([])
    #Get rid of all the information that is not a note
    if note.type == 'note_on' or note.type == 'note_off':
      time = time + note.time
      #In case that the full bar is in silence or there is a part of the bar in silence
      if time >= bar * bars and note.type=='note_on' and note.velocity > 0:
        if(len(barNotes[doneBars]) == 0):
          barNotes[doneBars].append(Message('note_on', note=0, velocity = 0, time = 0))
        extraTime = time
        #In case there are more than one bar in silence
        extraBar = False
        while extraTime >= bar * bars:
          if extraBar:
            codedSong.append({})
            barNotes.append([])
            codes.append([])
            doneBars = doneBars + 1
            barNotes[doneBars].append(Message('note_on', note=0, velocity = 0, time = 0))
          extraTime = extraTime - bar*bars
          #Control if before the full bar in silence was a bar not full
          if extraTime >= bar * bars and not extraBar:
            unfinished = doneBars
          extraBar = True
        barNotes[doneBars + 1].append(Message('note_on', note=note.note, velocity = 100, time = extraTime))
      #In case the last note is longer than the last bar ending, it is cutted to avoid overfitting the time note information
      elif time > bar * bars and (note.type=='note_off' or note.velocity == 0):
        barNotes[doneBars].append(Message('note_on', note=note.note, velocity = 0, time = note.time - (time - bar*bars)))
        barNotes[doneBars + 1].append(Message('note_on', note=note.note, velocity = 100, time = 0))
        barNotes[doneBars + 1].append(Message('note_off', note=note.note, velocity = 0, time = (time - bar*bars)))
        extraNote = [note.note, time - bar*bars]
        ending = True
      elif time <= bar * bars:
        barNotes[doneBars].append(note)
    #In case some control_effects take time
    else:
      time = time + note.time
    #If the next note is not inside the amount of bars we want to compute, it stops adding notes
    if (time >= bar * bars and (note.type == 'note_on' or note.type == 'note_off') or note.type == 'end_of_track') or unfinished != -1:
      if unfinished != -1:
        finalNotes, codes = CalculateFinalNotes(barNotes, unfinished, codes, finalNotes)
        codedSong[unfinished] = finalNotes
        doneBars = doneBars + 1
        codes.append([])
        codedSong.append({})
      else:
        finalNotes, codes = CalculateFinalNotes(barNotes, doneBars, codes, finalNotes)
        doneBars = doneBars + 1
        codedSong.append(finalNotes)
      finalNotes = {}
      #Make sure that its not a meta message and restore time
      if note.type == 'note_on' or note.type == 'note_off':
        if time > bar * bars and (note.type=='note_off' or note.velocity == 0):
          time = (time - bar*bars)
        elif time > bar * bars and len(barNotes[doneBars]) > 0:
          time = barNotes[doneBars][0].time
        else:
          time = 0
      else:
        time = 0
      #if doneBars == 6:
        #break
  #Making sure that there is not an extra note that is being deleted in case that the last note started in the previous bar
  if ending:
    codes.append([])
    codes[doneBars].append(extraNote[0])
    extraNotes = {}
    extraNotes[extraNote[0]] = extraNote[1]
    codedSong.append(extraNotes)
  print(codedSong)

  for j, finalNotes in enumerate(codedSong):
    #Once all the notes have their time played in the bar/s, the dictionary is sorted and the final Code is created with the 3 notes that have been played the most in the bar/s
    SortedNotes = sorted(finalNotes.items(), key=lambda x:x[1], reverse=True)
    #for debugging issues uncomment this to see the sorted notes by time of the bar/s
    #print(SortedNotes)
    #In case there are not even 3 musicalNotes in the bar
    if len(SortedNotes) == 1:
      codes[j].append(SortedNotes[0][0])
      codes[j].append(SortedNotes[0][0])
      codes[j].append(SortedNotes[0][0])
    elif len(SortedNotes) == 2:
      codes[j].append(SortedNotes[0][0])
      codes[j].append(SortedNotes[1][0])
      codes[j].append(SortedNotes[0][0])
    elif len(SortedNotes) >= 3:
      codes[j].append(SortedNotes[0][0])
      codes[j].append(SortedNotes[1][0])
      codes[j].append(SortedNotes[2][0])
    elif len(SortedNotes) == 0:
      codes[j].append(-1)
      codes[j].append(-1)
      codes[j].append(-1)
      codes[j].append(-1)
    #Finally, as the code uses only numbers from 0 to 11, each code is changed to this range
    for i, midiCode in enumerate(codes[j]):
      if midiCode > 11:
        codes[j][i] = midiCode % 12
    #Check if there is an extraCode
    if len(codes[-1]) == 0:
      codes.pop()

  #Make sure that the code is not an error because of overlapping notes
  for i, code in enumerate(codes):
    if len(code) > 4:
      print("There are two overlapping notes inbetween bar {} and bar {}".format(i, i+1))
      while len(code) > 4:
        code.pop()
      if code[0] != -1:
        for j in range(3, 0, -1):
          if code[j] == -1:
            code[j] = code[0]

  return codes
```

# **Database Creation Tools**

```python
def SetTonalityVariationsProgression (db):
  newDb = []
  #For each harmonic info create all the tonality variances
  for harmonicInfo in db:
    firstChord = harmonicInfo[0]
    actualTonality = harmonicInfo
    newDb.append(harmonicInfo)
    #Do the loop for the first time
    if actualTonality[5] != '-1':
      actualTonality = [NextChord(actualTonality[0]), actualTonality[1] + 1, actualTonality[2] + 1, actualTonality[3] + 1, actualTonality[4] + 1, NextChord(actualTonality[5])]
    else:
      actualTonality = [NextChord(actualTonality[0]), actualTonality[1] + 1, actualTonality[2] + 1, actualTonality[3] + 1, actualTonality[4] + 1, actualTonality[5]]
    for i, note in enumerate(actualTonality):
      if isinstance(note, int) and note > 11:
        actualTonality[i] = actualTonality[i] % 12
    newDb.append(actualTonality)
    #Loop for all the tonalities
    while newDb[-1][0] != firstChord:
      if actualTonality[5] != '-1':
        actualTonality = [NextChord(actualTonality[0]), actualTonality[1] + 1, actualTonality[2] + 1, actualTonality[3] + 1, actualTonality[4] + 1, NextChord(actualTonality[5])]
      else:
        actualTonality = [NextChord(actualTonality[0]), actualTonality[1] + 1, actualTonality[2] + 1, actualTonality[3] + 1, actualTonality[4] + 1, actualTonality[5]]
      for i, note in enumerate(actualTonality):
        if isinstance(note, int) and note > 11:
          actualTonality[i] = actualTonality[i] % 12
      newDb.append(actualTonality)
    #Take out the repeated first chord
    newDb.pop()

  return newDb
```

```python
def NextChord(chord):
  #Set the next chord posibility depending if it is a major or a minor chord
  majorChords = ['C', 'C#', 'D', 'D#', 'E', 'F', 'F#', 'G', 'G#', 'A', 'A#', 'B']
  minorChords = ['cm', 'c#m', 'dm', 'd#m', 'em', 'fm', 'f#m', 'gm', 'g#m', 'am', 'a#m', 'bm']
  if chord in majorChords:
    i = majorChords.index(chord)
    i = i + 1
    if i > 11:
      i = 0
    newChord = majorChords[i]
  elif chord in minorChords:
    i = minorChords.index(chord)
    i = i + 1
    if i > 11:
      i = 0
    newChord = minorChords[i]

  return newChord
```

```python
def DetectChords(code, lastChord):
  #Import the theory database
  import csv

  ChordNotes = {}
  with open('ChordNotes.csv', mode='r') as inp:
      reader = csv.reader(inp)
      ChordNotes = {rows[0]:[int(rows[1]), int(rows[2]), int(rows[3])] for rows in reader}
  ChordNotes['C'] = ChordNotes.pop('\ufeffC')

  #Import the Chords database
  y = []
  X = []
  possibleChords = ['C', 'C#', 'cm', 'c#m', 'D', 'D#', 'dm', 'd#m', 'E', 'em', 'F', 'F#', 'fm', 'f#m', 'G', 'G#', 'gm', 'g#m', 'A', 'A#', 'am', 'a#m', 'B', 'bm']
  with open('ProgressionChordsDataBase.csv', mode='r') as inp:
      reader = csv.reader(inp)
      for rows in reader:
        y.append(rows[0])
        X.append([int(rows[1]), int(rows[2]), int(rows[3]), int(rows[4]), rows[5]])
  for harmonicInfos in X:
    if harmonicInfos[4] in possibleChords:
      harmonicInfos[4] = possibleChords.index(harmonicInfos[4])
    else:
      harmonicInfos[4] = -1
  y[0] = 'C'

  #Append the last chord to the harmonic information to use chord progression
  from copy import deepcopy
  harmonicInfo = deepcopy(code)
  #OneHotEncoding for the last chord used
  if lastChord in possibleChords:
    lastChord = possibleChords.index(lastChord)
  else:
    lastChord = -1

  harmonicInfo[0].append(lastChord)

  
  #Fit the knn with the db
  from sklearn.neighbors import KNeighborsClassifier
  neigh = KNeighborsClassifier(n_neighbors=5)
  neigh.fit(X, y)

  #Getting the punctuation per chords
  predictedChord = []
  punctuation = []
  #print(neigh.classes_)
  #print(neigh.predict_proba(harmonicInfo)[0])
  for i, prob in enumerate(neigh.predict_proba(harmonicInfo)[0]):
    if prob > 0:
      #Adding a possible chord that is at least predicted by the Knn
      predictedChord.append(neigh.classes_[i])
      punctuation.append(0)
      #Adding score to the chord for each note that is part of it in the bar
      if harmonicInfo[0][0] in chordNotes[neigh.classes_[i]]:
        punctuation[-1] = punctuation[-1] + 1
      elif harmonicInfo[0][0] in fourthChordNotes[neigh.classes_[i]]:
        punctuation[-1] = punctuation[-1] + 0.5
      if harmonicInfo[0][1] in chordNotes[neigh.classes_[i]]:
        punctuation[-1] = punctuation[-1] + 1
      elif harmonicInfo[0][1] in fourthChordNotes[neigh.classes_[i]]:
        punctuation[-1] = punctuation[-1] + 0.5
      if harmonicInfo[0][2] in chordNotes[neigh.classes_[i]]:
        punctuation[-1] = punctuation[-1] + 1
      elif harmonicInfo[0][2] in fourthChordNotes[neigh.classes_[i]]:
        punctuation[-1] = punctuation[-1] + 0.5
      if harmonicInfo[0][3] in chordNotes[neigh.classes_[i]]:
        punctuation[-1] = punctuation[-1] + 1
      elif harmonicInfo[0][3] in fourthChordNotes[neigh.classes_[i]]:
        punctuation[-1] = punctuation[-1] + 0.5

  #Sort chords by preference
  sortedChords = SortChords(punctuation, predictedChord)
  #for i, chords in enumerate(predictedChord):
    #print("{} has a punctuation of {}".format(chords, punctuation[i]))

  #print("\nThe most probable chord is {}".format(predictedChord[punctuation.index(max(punctuation))]))
  #Getting the probability just using the theory
  theoryProbs = {}
  for possibleChord in chordNotes:
    theoryProbs[possibleChord] = 0
    if harmonicInfo[0][0] in chordNotes[possibleChord]:
      theoryProbs[possibleChord] = theoryProbs[possibleChord] + 1
    elif harmonicInfo[0][0] in fourthChordNotes[possibleChord]:
      theoryProbs[possibleChord] = theoryProbs[possibleChord] + 0.5
    if harmonicInfo[0][1] in chordNotes[possibleChord]:
      theoryProbs[possibleChord] = theoryProbs[possibleChord] + 1
    elif harmonicInfo[0][1] in fourthChordNotes[possibleChord]:
      theoryProbs[possibleChord] = theoryProbs[possibleChord] + 0.5
    if harmonicInfo[0][2] in chordNotes[possibleChord]:
      theoryProbs[possibleChord] = theoryProbs[possibleChord] + 1
    elif harmonicInfo[0][2] in fourthChordNotes[possibleChord]:
      theoryProbs[possibleChord] = theoryProbs[possibleChord] + 0.5
    if harmonicInfo[0][3] in chordNotes[possibleChord]:
      theoryProbs[possibleChord] = theoryProbs[possibleChord] + 1
    elif harmonicInfo[0][3] in fourthChordNotes[possibleChord]:
      theoryProbs[possibleChord] = theoryProbs[possibleChord] + 0.5

  theoryProbs = {k: v for k, v in sorted(theoryProbs.items(), key=lambda item: item[1], reverse=True)}

  #print(theoryProbs)
  return [sortedChords, theoryProbs]
```

```python
def SortChords(punctuation, predictedChord):
  sortedChords = []
  for n in range(8,0,-1):
    for i, score in enumerate(punctuation):
      if score == n/2:
        sortedChords.append(predictedChord[i])

  return sortedChords
```

```python
def DetectChordsError(code, lastChord,k=4):
  #Append the last chord to the harmonic information to use chord progression
  from copy import deepcopy
  harmonicInfo = deepcopy(code)

  harmonicInfo[0].append(lastChord)

  
  #Fit the knn with the db
  from sklearn.neighbors import KNeighborsClassifier
  neigh = KNeighborsClassifier(n_neighbors=k)
  neigh.fit(X, y)

  #Getting the punctuation per chords
  predictedChord = []
  punctuation = []
  for i, prob in enumerate(neigh.predict_proba(harmonicInfo)[0]):
    if prob > 0:
      #Adding a possible chord that is at least predicted by the Knn
      predictedChord.append(neigh.classes_[i])
      punctuation.append(0)
      #Adding score to the chord for each note that is part of it in the bar
      if harmonicInfo[0][0] in chordNotes[neigh.classes_[i]]:
        punctuation[-1] = punctuation[-1] + 1
      elif harmonicInfo[0][0] in fourthChordNotes[neigh.classes_[i]]:
        punctuation[-1] = punctuation[-1] + 0.5
      if harmonicInfo[0][1] in chordNotes[neigh.classes_[i]]:
        punctuation[-1] = punctuation[-1] + 1
      elif harmonicInfo[0][1] in fourthChordNotes[neigh.classes_[i]]:
        punctuation[-1] = punctuation[-1] + 0.5
      if harmonicInfo[0][2] in chordNotes[neigh.classes_[i]]:
        punctuation[-1] = punctuation[-1] + 1
      elif harmonicInfo[0][2] in fourthChordNotes[neigh.classes_[i]]:
        punctuation[-1] = punctuation[-1] + 0.5
      if harmonicInfo[0][3] in chordNotes[neigh.classes_[i]]:
        punctuation[-1] = punctuation[-1] + 1
      elif harmonicInfo[0][3] in fourthChordNotes[neigh.classes_[i]]:
        punctuation[-1] = punctuation[-1] + 0.5

  #Sort chords by preference
  sortedChords = SortChords(punctuation, predictedChord)
  
  #Getting the probability just using the theory
  theoryProbs = {}
  for possibleChord in chordNotes:
    theoryProbs[possibleChord] = 0
    if harmonicInfo[0][0] in chordNotes[possibleChord]:
      theoryProbs[possibleChord] = theoryProbs[possibleChord] + 1
    elif harmonicInfo[0][0] in fourthChordNotes[possibleChord]:
      theoryProbs[possibleChord] = theoryProbs[possibleChord] + 0.5
    if harmonicInfo[0][1] in chordNotes[possibleChord]:
      theoryProbs[possibleChord] = theoryProbs[possibleChord] + 1
    elif harmonicInfo[0][1] in fourthChordNotes[possibleChord]:
      theoryProbs[possibleChord] = theoryProbs[possibleChord] + 0.5
    if harmonicInfo[0][2] in chordNotes[possibleChord]:
      theoryProbs[possibleChord] = theoryProbs[possibleChord] + 1
    elif harmonicInfo[0][2] in fourthChordNotes[possibleChord]:
      theoryProbs[possibleChord] = theoryProbs[possibleChord] + 0.5
    if harmonicInfo[0][3] in chordNotes[possibleChord]:
      theoryProbs[possibleChord] = theoryProbs[possibleChord] + 1
    elif harmonicInfo[0][3] in fourthChordNotes[possibleChord]:
      theoryProbs[possibleChord] = theoryProbs[possibleChord] + 0.5

  theoryProbs = {k: v for k, v in sorted(theoryProbs.items(), key=lambda item: item[1], reverse=True)}

  #Get the score of each one of the sorted chords
  sortedPunctuations = []
  for chord in sortedChords:
    sortedPunctuations.append(theoryProbs[chord])

  #print(theoryProbs)
  return [sortedChords, sortedPunctuations]
```

```python
#Import the theory database
import csv

ChordNotes = {}
with open('ChordNotes.csv', mode='r') as inp:
    reader = csv.reader(inp)
    ChordNotes = {rows[0]:[int(rows[1]), int(rows[2]), int(rows[3])] for rows in reader}
ChordNotes['C'] = ChordNotes.pop('\ufeffC')
```

```python
#Import the Chords database
y = []
X = []
possibleChords = ['C', 'C#', 'cm', 'c#m', 'D', 'D#', 'dm', 'd#m', 'E', 'em', 'F', 'F#', 'fm', 'f#m', 'G', 'G#', 'gm', 'g#m', 'A', 'A#', 'am', 'a#m', 'B', 'bm']
with open('ProgressionChordsDataBase.csv', mode='r') as inp:
    reader = csv.reader(inp)
    for rows in reader:
      y.append(rows[0])
      X.append([int(rows[1]), int(rows[2]), int(rows[3]), int(rows[4]), rows[5]])
for harmonicInfos in X:
  if harmonicInfos[4] in possibleChords:
    harmonicInfos[4] = possibleChords.index(harmonicInfos[4])
  else:
    harmonicInfos[4] = -1
y[0] = 'C'
```
